```python
class Registration():

    def __init__(self):
        self.menu()
        
    def menu(self):
        import sys
        menu_input = input('''
        press 1 to Register
        press 2 to Login
        press 3 to exit
        ''')
#getting the input from user for user ID        
        if menu_input=='1':
            self.user_validation()
        elif menu_input=='2':
            self.login() #Callin the login class
        else:
            print("Stopped")
            return
#-------------x------------(Usesr ID validation)----------------x--------------x
    def user_validation(self):
        
        self.reg_userid = input('Enter the User ID')# Getting the input for new user ID from theh user

        flag1 = False

        # Checking the user name contains '@', '.' in the user input
        if '@' in self.reg_userid and '.' in self.reg_userid:
            flag1 = True  
        else: 
            self.menu()
        #Checking for required format of User ID 
        #--------x------(nothing@domain.com)---------x
        if flag1:
            import re,os,json
            ind = []
            index1 = re.finditer('[@.]',self.reg_userid)
            for i in index1:
                ind.append(i.start())#appending the index value of '@' and  '.' in the user input
            start = len(self.reg_userid[:ind[0]])
            betw = len(self.reg_userid[ind[0]:ind[1]])

            #checking length '@' and  '.' in the user input to meet the required format
            if start>=3 and betw>=3:
                match1 =re.search('^[0-9]',self.reg_userid)
                match2 = re.search('^[~`!@#$%^&*()-_=+-,.|\\\\/]',self.reg_userid)

                if match1:
                    print()
                    print('User ID should not start with any numbers')
                    self.user_validation()
                elif match2:
                    print()
                    print('User ID should not start with any special charectors\nor any capital letter')
                    self.user_validation()
                else:
                    print()
                    print('User ID requirement satisfied')
                    self.password_validation()# Calling function for password validation
            else:
                print()
                print("Requirement not met,\nPlease create user id like 'abcd@domain.com'")
                self.user_validation()
        else:
            print("\nRequirement not met,\nPlease create user id like 'abcd@domain.com'")
            self.user_validation()                    
#-------------x----------------(Password validation)---------------x-------------------------x    
    def password_validation(self):
        import re
        self.reg_password = input("Set password")
        flag2 = False
        if len(self.reg_password)>5 and len(self.reg_password)<16:
            match3 = re.search('[a-z]',self.reg_password)
            if match3:
                match4 = re.search('[A-Z]',self.reg_password)
                if match4:
                    match5 = re.search('[0-9]',self.reg_password)
                    if match5:
                        special = ['~','`','!','@','#','$','%','^','&','*','(',')','-','_','+','=','.',',','\\','/']
                        match7 = any(char in special for char in self.reg_password)
                        if match7:
                            flag2 = True
                        else: 
                            print("Password should contain atlest one special charector")
                            self.password_validation()
                    else: 
                        print("Password should contain atlest one numeric")
                        self.password_validation()
                else: 
                    print("Password should contain atlest one capital letter")
                    self.password_validation()
            else: 
                print("Password should contain atlest one small letter")
                self.password_validation()
        else:
            print("Password ength should be between 5 to 16")
            self.password_validation()
#---------x---------(Writing the file)-------------x------------------x            
        if flag2:
          import json, os
          file_name =  'Register.json'
          if os.path.exists(file_name):
              with open(file_name,'r') as f:
                  employee = json.load(f)
          else:
              employee = {"user":[],"password":[]}

          while True:
              employee["user"].extend([self.reg_userid])
              employee["password"].extend([self.reg_password])
              with open(file_name,'w') as wf:
                json.dump(employee,wf)
                print('Registration Successfull')
              break
        else:
            print("Password requirements not matched, please try again")

#-------x----------x-(Login)--------x-----------x--------------x-----------x

    class login:
        
        def __init__(self):
            self.user_login() # Callin the login option by default
        def user_login(self):
            import os
            try:
                os.path.exists('Register.json')
                import time
                self.registred_user_id = input("Enter the User ID to Login:")
                self.user_password = input("Enter the Password:")
                import json
                with open('Register.json','r') as f:
                    login_check = json.load(f)
                    if self.registred_user_id in login_check['user'] and self.user_password in login_check['password']:
                        print("Loggin successfull")
                        return
                    else:
                        import time
                        time.sleep(0.5)
                        print('\nInvalid "User ID or Password"')
                        menu_input = input("""
                        Press 1 to register
                        Press 2 forget password
                        Press 3 to exit""")

                        if menu_input =='1':
                            print('Please Signup')
                            time.sleep(0.5)
                            signup = Registration()
                            signup.menu()
                        elif menu_input =='2':
                            time.sleep(0.5)
                            self.forget() # Calling forget password function
                        else:
                            print("'Stopped'")
                            return 
            except:
                print("Please Signup")
                signup = Registration()
                signup.menu()
#--------x------------x(Forget Password)x-----------------x------------x
        class forget:
            def __init__(self):
                self.forget_password_menu()
            def forget_password_menu(self):
                    import time
                    forget_menu = input('''
                    Press 1 to Retrive password
                    Press 2 to Create New password
                    Press 3 for Registration
                    Press 4 for Exit 
                    ''')

                    if forget_menu == '1':
                        #calling  password retrive function function
                        time.sleep(0.5)
                        self.retrive()  
                    elif forget_menu == '2':
                        #call create new password function function
                        time.sleep(0.5)
                        self.new_password()
                    elif forget_menu == '3':
                        time.sleep(0.5)
                        print("\nPlease Sign up")
                        signup = Registration()
                    else:
                        print("Stopped")
                        return 

#--------x------------x------(Retriving Password)------------x---------------x
            def retrive(self):
                import json
                user_input = input('Enter Registered user ID')
                with open('Register.json','r') as f:
                    find_password = json.load(f)
                    import time
                    if user_input in find_password['user']:
                        ind = find_password['user'].index(user_input)
                        time.sleep(0.5)
                        print(f"\nPassword for user ID '{user_input}' is:\n{find_password['password'][ind]}")
                    else:
                        print(f'\nNo such "{user_input}" user ID found\nPlease Create New User ID')
                        time.sleep(0.5)
                        signup = Registration() 
                        signup.menu()
                        
#x----------x--------------x-(Creating New Password)------------x---------------x
            def new_password(self):
                    flag5 = False
                    import json
                    with open('Register.json','r') as rf:
                        self.old_userID1 = input('Enter user id to reset password')
                        update_password = json.load(rf)
                        if self.old_userID1 in update_password['user']:
                            flag5 = True 
                        else:
                            flag5 = False    
                    if flag5:
                        #calling function for validation of new password
                        self.new_password_validation()                                                                                                                       
                    else:
                        print('\nNo such User ID found\nPlease Sign Up')
 #-------------x--------x(New Password validation)x---------x----------x                       
            def new_password_validation(self):
                import re,time,json
                time.sleep(0.5)
                new_password = input("Set password")
                if len(new_password)>5 and len(new_password)<16:
                    match3 = re.search('[a-z]',new_password)
                    if match3:
                        match4 = re.search('[A-Z]',new_password)
                        if match4:
                            match5 = re.search('[0-9]',new_password)
                            if match5:
                                special = ['~','`','!','@','#','$','%','^','&','*','(',')','-','_','+','=','.',',','\\','/']
                                match7 = any(char in special for char in new_password)
                                if match7:
                                    print('update the password')
                                    #call the update_password()
                                    with open("Register.json", 'r') as rf:
                                        update_password = json.load(rf)
                                        while True:
                                            ind = update_password['user'].index(self.old_userID1)
                                            update_password['password'][ind] = new_password 
                                            with open('Register.json','w') as wf:
                                                json.dump(update_password,wf)
                                                #dumping the data
                                                print(f"\nPassword Updated for User ID {self.old_userID1}")
                                            break
                                else:
                                    print("Password should contain atlest one special charector")
                                    self.new_password_validation()   
                            else: 
                                print("Password should contain atlest one numeric")
                                self.new_password_validation()                                                
                        else: 
                            print("Password should contain atlest one capital letter")
                            self.new_password_validation()
                    else:
                        print("Password should contain atlest one small letter")
                        self.new_password_validation()
                else:
                    print("Password ength should be between 5 to 16")
                    self.new_password_validation()
                          
```
